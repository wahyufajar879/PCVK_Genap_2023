### **2.   Import Librari**

```python
# Import beberapa library

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 3 dilasi pada citra dengan Structuring Element Dilasi (SED) berbentuk cross 3 x 3 
# tanpa menggunakan library morphology

from google.colab.patches import cv2_imshow
img_morph = cv2.imread('/content/drive/MyDrive/PCVK/Week9/Image/morphology.png')

mask = np.zeros((3,3), dtype=np.uint8)
mask[1, :] = 1  # garis horizontal
mask[:, 1] = 1  # garis vertikal

img_dilation = cv2.dilate(img_morph, mask, iterations=1)

plt.subplot(121),plt.imshow(cv2.cvtColor(img_morph, cv2.COLOR_BGR2RGB)), plt.title('Gambar Asli')
plt.subplot(122),plt.imshow(cv2.cvtColor(img_dilation, cv2.COLOR_BGR2RGB)), plt.title('Gambar Dilasi')
plt.show()
```

```python
# 4 dilasi Structuring Element Dilasi (SED) berbentuk cross 3 x 3 
# dengan menggunakan library morphology dari openCV

mask = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))

# Perform dilation on the image
img_dilation = cv2.dilate(img_morph, mask, iterations=1)

# Display the original and dilated images side by side
plt.subplot(121), plt.imshow(cv2.cvtColor(img_morph, cv2.COLOR_BGR2RGB)), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cv2.cvtColor(img_dilation, cv2.COLOR_BGR2RGB)), plt.title('Citra Dilasi')
plt.show()

```

```python
# 5 hasil erosi pada citra dengan Structuring Element Erosi (SE) berbentuk square 5 x 5 
# tanpa menggunakan library morphology dari openCV

from google.colab.patches import cv2_imshow
img_j = cv2.imread('/content/drive/MyDrive/PCVK/Week9/Image/j.png')

mask = np.ones((5,5), np.uint8)
img_erosi = cv2.erode(img_j, mask, iterations=1)

plt.subplot(121),plt.imshow(img_j), plt.title('Citra Awal')
plt.subplot(122),plt.imshow(img_erosi), plt.title('Citra Hasil Erosi')
```

```python
# 6 dilasi Structuring Element Erosi (SE) berbentuk square 5 x 5 
# dengan menggunakan library morphology dari openCV
mask = np.ones((5,5), np.uint8)
morph_erosi = cv2.morphologyEx(img_j, cv2.MORPH_ERODE, mask)

plt.subplot(121),plt.imshow(img_j), plt.title('Citra Awal')
plt.subplot(122),plt.imshow(morph_erosi), plt.title('Citra Hasil Erosi')
```

```python
# 7 operasi opening dengan Structuring Element berbentuk square 3 x 3 
# tanpa menggunakan library morphology dari openCV

from google.colab.patches import cv2_imshow
img_finger = cv2.imread('/content/drive/MyDrive/PCVK/Week9/Image/fingerprint.png')

mask = np.ones((3,3), np.uint8)
erosi = cv2.erode(img_finger, mask, iterations=1)
opening = cv2.dilate(erosi, mask, iterations=1)

plt.subplot(121),plt.imshow(mask), plt.title('Citra Awal')
plt.subplot(121),plt.imshow(erosi), plt.title('Citra Hasil Erosi')
plt.subplot(122),plt.imshow(opening), plt.title('Citra Hasil Opening')
```

```python
# 7 operasi opening dengan Structuring Element berbentuk square 3 x 3 
# dengan menggunakan library morphology dari openCV

mask = np.ones((3,3), np.uint8)
erosi = cv2.erode(img_finger, mask, iterations=1)
morph_opening = cv2.morphologyEx(img_finger, cv2.MORPH_OPEN, mask)

plt.subplot(121),plt.imshow(img_finger), plt.title('Citra Awal')
plt.subplot(121),plt.imshow(erosi), plt.title('Citra Hasil Erosi')
plt.subplot(122),plt.imshow(morph_opening), plt.title('Citra Hasil Opening')
```

```python
# 8 operasi closing dengan Structuring Element berbentuk square 3 x 3 
# tanpa menggunakan library morphology dari openCV

mask = np.ones((3,3), np.uint8)
dilasi = cv2.dilate(img_finger, mask, iterations=1)
closing = cv2.erode(dilasi, mask, iterations=1)

plt.subplot(121),plt.imshow(img_finger), plt.title('Citra Awal')
plt.subplot(121),plt.imshow(dilasi), plt.title('Citra Hasil Dilasi')
plt.subplot(122),plt.imshow(closing), plt.title('Citra Hasil Closing')
```

```python
# 8 operasi closing dengan Structuring Element berbentuk square 3 x 3
# dengan menggunakan library morphology dari openCV

mask = np.ones((3,3), np.uint8)
dilasi = cv2.dilate(img_finger, mask, iterations=1)
morph_closing = cv2.morphologyEx(img_finger, cv2.MORPH_CLOSE, mask)

plt.subplot(121),plt.imshow(img_finger), plt.title('Citra Awal')
plt.subplot(121),plt.imshow(dilasi), plt.title('Citra Hasil Dilasi')
plt.subplot(122),plt.imshow(morph_closing), plt.title('Citra Hasil Closing')
```
